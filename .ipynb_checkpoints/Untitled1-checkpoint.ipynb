```python
import pandas as pd
```

```python
# Emma
emma = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vQ-nAsuxYqX5Fr0b53FtKfHNmtQM2g-YOD5vpqrWKb4CSL8wAXSBdi2BgCZQ_eBT4J-gT-G0d3RSMB6/pub?gid=68869620&single=true&output=csv' 
bleakhouse = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vT1DJPuI92uL4BHBCI25XEI4oU7dO1AIgVWWypM8k-aoxcNKqg7IIKg02o1h4vw7bB9KKMMj2TcMcFO/pub?gid=1649677677&single=true&output=csv'
```

```python
df=pd.read_csv(bleakhouse).fillna('')
```

```python
df
```

```python
nameld=[]
gby='name_real'
other_cols='gender	race	class	other	notes'.split()
for name,namedf in df.groupby(gby):
    namedx={}
    namedx['ID']=name
    namedx['Label']=name
    namedx['Names'] = ', '.join(namedf.names)
    namedx['Num'] = sum(namedf.num)
    for oc in other_cols: namedx[oc]=namedf[oc].iloc[0]
    nameld.append(namedx)
newdf=pd.DataFrame(nameld).sort_values('Num',ascending=False)
newdf['Rank']=[i+1 for i,x in enumerate(newdf.index)]
newdf
```

```python
newdf.to_csv('data.chars.bleakhouse.csv')
```
